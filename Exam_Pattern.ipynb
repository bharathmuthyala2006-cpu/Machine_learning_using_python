# Feature Engineering

Create Activity Category

---

Teacher Says:

People walking:

```text
Less than 7000
=
Inactive

7000 - 10000
=
Moderately Active

Above 10000
=
Highly Active
```

---

Example

```python
def activity(steps):

    if steps > 10000:
        return "Highly Active"

    elif steps >= 7000:
        return "Moderately Active"

    else:
        return "Inactive"
```

---

```python
df["Activity_Level"] = (
df["Daily_Steps"]
.apply(activity)
)
```

---

## Count Activity Levels

```python
df["Activity_Level"].value_counts()
```
