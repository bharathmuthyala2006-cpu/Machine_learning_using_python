Create a simple ATM PIN system.
Correct PIN = 1234
Allow the user a maximum of 3 attempts.
If the PIN is correct:
"Transaction Allowed"
Otherwise:
"Card Blocked"
Hint:
Use a counter variable.
Use a while loop.
Increase the counter after every wrong attempt.


```python
correct_pin = "1234"

attempts = 0

while attempts <=3:
  pin = input("Enter the PIN: ")

  if pin == correct_pin:
    print("Transaction Allowed")

  else:
    print("Card Blocked")

  attempts +=1
```

```python
correct_pin = "1234"
attempts_pin = 0


while attempts_pin < 3:
  pin = input("Enter the PIN: ")
  if pin == correct_pin:
    print("Transaction Allowed")
    break
  else:
    print("Wrong PIN number")
    attempts_pin += 1

if attempts_pin == 3:
  print("Card is Blocked")
```
